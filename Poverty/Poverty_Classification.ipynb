# Data Preprocessing

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
from sklearn.preprocessing import StandardScaler, LabelBinarizer, LabelEncoder
from sklearn.svm import SVC
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
#from sklearn.pipeline import Pipeline
#from sklearn_pandas import DataFrameMapper
```

```python
def preprocess_data(data):
    print('Shape before: {}'.format(data.shape))
    data = data.drop('id', axis=1)
    for col in data.columns:
        if data.dtypes[col] == 'object':
            le = LabelBinarizer()
            le.fit(data[col])
            data[col] = le.transform(data[col])
    
        #elif data.dtypes[col] == 'int64':
         #   scaler = StandardScaler()
          #  temp = data[col].values.reshape(-1, 1)
           # scaler.fit(temp)
            #data[col] = scaler.transform(temp)
    print('Shape after: {}:'.format(data.shape))
    return data
```

I initially tried standardizing all the numerical categories, but I was actually getting worse results when I was doing so. 

```python
poverty_train_transformed = preprocess_data(poverty_train)
```

```python
poverty_test_transformed = preprocess_data(poverty_test)
```

```python
# Change the labels to 0 and 1
poverty_train_transformed['poor'] = poverty_train_transformed.poor.astype('int')
poverty_test_transformed['poor'] = poverty_test_transformed.poor.astype('int')
```

```python
X_train = poverty_train_transformed.drop(['poor', 'country'], axis=1)
X_test = poverty_test_transformed.drop(['poor', 'country'], axis=1)
y_train = poverty_train_transformed['poor']
y_test = poverty_test_transformed['poor']
```

# Initial Modeling and Getting Feature Importances

My initial thought is to run a logistic regression and see what my log loss is.

```python
log_r = LogisticRegression(penalty='l1')
log_r.fit(X_train, y_train)
y_pred = log_r.predict_proba(X_test)
log_loss(y_test, y_pred)
```

Now that I have an initial baseline metric I want to improve, I use Random Forest to grab the important features and decide which ones I want to remove.

```python
rfc = RandomForestClassifier(n_estimators=50, max_depth=10)
rfc.fit(X_train, y_train)
y_pred = rfc.predict_proba(X_test)
log_loss(y_test, y_pred)
```

```python
feature_importances = pd.DataFrame(rfc.feature_importances_,
                                   index = X_train.columns,
                                   columns=['importance']).sort_values('importance', ascending=False)

unimportant_features = feature_importances[feature_importances['importance'] < 0.000001].index
```

```python
poverty_train_transformed.drop(unimportant_features, axis=1, inplace=True)
poverty_test_transformed.drop(unimportant_features, axis=1, inplace=True)
```

```python
poverty_train_transformed.shape
```

After determining some of the unimportant features, I have removed about 20 features.  Now let's check to see if the models performed better.

# Check for Improvements

```python
X_train = poverty_train_transformed.drop(['poor', 'country'], axis=1)
X_test = poverty_test_transformed.drop(['poor', 'country'], axis=1)
y_train = poverty_train_transformed['poor']
y_test = poverty_test_transformed['poor']

log_r = LogisticRegression(penalty='l1', solver='liblinear')
log_r.fit(X_train, y_train)
y_pred = log_r.predict_proba(X_test)
log_loss(y_test, y_pred)
```

I can see that my Logistic Regression log loss improved from about ~0.31 to ~0.27.

```python
svc = SVC(C=1, probability=True, verbose=True)
svc.fit(X_train, y_train)
y_pred = svc.predict_proba(X_test)
log_loss(y_test, y_pred)
```

Using a Support Vector Classifier model gives me about the same result.
